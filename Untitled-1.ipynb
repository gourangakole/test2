```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
def f1(x):
    return (15*x**3 + 21*x**2 + 41*x + 3)**(1/4)*(np.exp(-0.5*x))
def f2(x):
    return (1.0/x**2)
def f3(x):
    return (-5.0/x)
def f4(x, R):
    return (np.exp(-x/R)/x)
```

```python
x = np.linspace(0.5, 10, 50)
plt.plot(x, f1(x), color='red',label="f1(x)")
plt.plot(x, f2(x), linestyle='dashed', color='green', label="f2(x)")
plt.xlabel("x",fontsize="15")
plt.ylabel("$f^2$",fontsize="15")
plt.legend(loc="best", fontsize="15")
plt.grid()
plt.title("Mutiple plots")
plt.show()
# plt.savefig("multiple.pdf")
print ("Plotting is Done!")
```
